```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import SimpleITK as sitk
import matplotlib.pyplot as plt

sns.set(context="paper", style="white", font_scale=3.5, rc={'axes.linewidth': 0.05})
```

### Dataset Loading

```python
T_1, T_2, T_3, T_4, T_5, T_6, T_7 = list(), list(), list(), list(), list(), list(), list()
paths = [
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/A_RUNMC_2l',
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/B_BMC_2l',
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/C_I2CVB_HCRUDB_1l',
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/D_UCL_1l',
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/E_BIDMC_1l',
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/F_HK_1l',
         '/home/aranem_locale/Desktop/mnts/local/scratch/aranem/MICCAI_2023/MICCAI_2023__raw_data/raw/Sources/Transformed/Task05_Prostate/imagesTr'
        ]

# Load data (use Transformed ones as they have all the same shape!
for p, data in zip(paths, [T_1, T_2, T_3, T_4, T_5, T_6, T_7]):
    list_of_imgs = [x for x in os.listdir(p) if '._' not in x and '.json' not in x and 'DS_Store' not in x and 'egmentation' not in x]
    # print(len(list_of_imgs))
    for img in list_of_imgs:
        img_ = sitk.GetArrayFromImage(sitk.ReadImage(os.path.join(p, img)))
        img_ = (img_ - img_.min()) / (img_.max() - img_.min())  # Normalize between 0 and 1
        data.append(img_)

        
T_1 = pd.DataFrame(np.concatenate(T_1).ravel().tolist(), columns=['Intensities'])
T_1['ID'] = '$\mathcal{T}_{1}$'
T_2 = pd.DataFrame(np.concatenate(T_2).ravel().tolist(), columns=['Intensities'])
T_2['ID'] = '$\mathcal{T}_{2}$'
T_3 = pd.DataFrame(np.concatenate(T_3).ravel().tolist(), columns=['Intensities'])
T_3['ID'] = '$\mathcal{T}_{3}$'
T_4 = pd.DataFrame(np.concatenate(T_4).ravel().tolist(), columns=['Intensities'])
T_4['ID'] = '$\mathcal{T}_{4}$'
T_5 = pd.DataFrame(np.concatenate(T_5).ravel().tolist(), columns=['Intensities'])
T_5['ID'] = '$\mathcal{T}_{5}$'
T_6 = pd.DataFrame(np.concatenate(T_6).ravel().tolist(), columns=['Intensities'])
T_6['ID'] = '$\mathcal{T}_{6}$'
T_7 = pd.DataFrame(np.concatenate(T_7).ravel().tolist(), columns=['Intensities'])
T_7['ID'] = '$\mathcal{T}_{7}$'

# dfList = [T_1, T_2, T_3, T_4, T_5, T_6, T_7]
# data = pd.concat(dfList)
```
